## ***3. Data Wrangling***

### Data Wrangling Code

```python
# Creating a copy of original dataframe
new_df = df.copy()
# Adding a new column customer id 
new_df.insert(0,'Customer_id',list(range(1,3334)))
```

```python
# Adding a new column Average call duration
new_df['avg_calls_duration'] = (new_df['Total night minutes'] + new_df["Total day minutes"]+new_df["Total eve minutes"])/(new_df['Total night calls']+new_df['Total day calls']+new_df['Total eve calls'])
```

```python
# Sorting out customers on the basis of duration of call
def call_type(length):
  if length > 2.5:
    return 'long'
  elif length > 1.9:
    return 'medium'
  else:
    return 'short'

new_df['Call_duration'] = new_df['avg_calls_duration'].apply(lambda x: call_type(x))
```

```python
# Sorting out customers on the basis of no. days of subscription.

def customer_type(length):
  if length >=120:
    return 'old'
  elif length > 75:
    return 'intermediate'
  else:
    return 'new'

# Adding a new column Customer type based on the above separation
new_df['customer_type']=new_df['Account length'].apply(customer_type)
```

```python
# Checking first 5 rows of our dataframe
new_df.head()
```

```python
# Grouping by Churn & calculating mean values for rest of the columns.
churn_count = df.groupby('Churn')
churn_count.mean()
```

```python
copy_df = df.copy()

def convert_to_bool(elem):
  
  if elem == 'Yes':
    return True
  else:
    return False

copy_df['International_plan_bool'] = copy_df['International plan'].apply(lambda x: convert_to_bool(x))
copy_df['Voice_mail_plan_bool'] = copy_df['Voice mail plan'].apply(lambda x: convert_to_bool(x))
```

### What all manipulations have you done and insights you found?

1. Added a new column as customer id as values present in it are serial numbers from 1 to 3333.
2. More number of customer services have been made among the churned customers.
3. Added a new column as "avg_calls_duration" which is average duration of total calls &	"Call_duration" on the basis of duration of call, criteria for sorting is
                     ----> Above 2.5, Long duration
                     ----> In between 1.9 to 2.5, Medium duration
                     ----> Below 1.9, Short duration

4. Added a new column as "customer_type" on the basis of days of subscription, criteria for sorting is
                     ----> Above 120, old customer 
                     ----> In between 75 & 120, intermediate
                     ----> Below 75, new
5. We have created a copy of dataframe and converted international plan and voice mail plans into boolean values to find corelations.

6. Outliers related to customer account length is ignored by taking them as old customers as that can be a possible scenario.

7. Similarly outliers in customer service calls are ignored as there might be more number of calls due to poorer service.

### Data Wrangling Code Continuation

```python
#Comparing customer types with churn
customer_type_vs_churn= new_df.groupby(['customer_type'])["Churn"].value_counts().unstack()
customer_type_vs_churn['total customer'] = customer_type_vs_churn[0]+customer_type_vs_churn[1]
customer_type_vs_churn['churn percentage'] = round(100*(customer_type_vs_churn[1] / customer_type_vs_churn['total customer']),2)
customer_type_vs_churn
```

```python
#Checking customer distribution in states on basis of customer type
customer_type1=new_df.groupby('State')['customer_type'].value_counts().unstack()
customer_type1 = customer_type1.reset_index().rename(columns = {"State":"States"})
statewise_customers = new_df['State'].value_counts().reset_index().rename(columns = {'index':'States','State':'No. of customers'})
state_wise_type_of_C = pd.merge(customer_type1,statewise_customers,on='States')
state_wise_type_of_C.head()
```

```python
# Getting statewise churn percentage.
users_statewise_churn=new_df.groupby('State')['Churn'].value_counts().unstack()
users_statewise_churn['total_customer'] = users_statewise_churn[0]+users_statewise_churn[1]
users_statewise_churn['churn percentage'] = round(100*(users_statewise_churn[1]/users_statewise_churn['total_customer']),2)
users_statewise_churn.sort_values(by = 'churn percentage',ascending = False,inplace = True)
users_statewise_churn = users_statewise_churn.reset_index()
users_statewise_churn.rename(columns ={False:'churn_False',True:'churn_True'},inplace = True)
users_statewise_churn.head()
```

```python
# International plans subscribers
international_plans = new_df['International plan'].value_counts().reset_index().rename(columns = {'index':'International_plan_opted','International plan':'Subscribers'})
international_plans['percentage'] = international_plans['Subscribers'].apply(lambda x : round((x/3333)*100,2))
international_plans 
```

```python
# Voice Mail plans subscribers
voicemail_plans = new_df['Voice mail plan'].value_counts().reset_index().rename(columns = {'index':'Vmail_plan_opted','Voice mail plan':'Subscribers'})
voicemail_plans['percentage'] = voicemail_plans['Subscribers'].apply(lambda x : round((x/3333)*100,2))
voicemail_plans
```

```python
# Customers who subscribed both plans
persons_who_subscribe_both_plans = new_df.groupby(['Voice mail plan','International plan'])["Customer_id"].count()
persons_who_subscribe_both_plans.unstack()
```

```python
# Finding out relation between number of minutes & calls for customers with international plans
inter_calls_churn = new_df[(new_df['International plan'] == 'Yes') & (new_df["Total intl calls"] <= 5)][['Total intl calls','Total intl minutes','Churn']]
inter_calls_churn.sort_values(by ='Total intl minutes', ignore_index = True,inplace = True)
inter_calls_churn
```

```python
#Churn Distribution on international callers with less number of calls
inter_calls_churn = inter_calls_churn['Churn'].value_counts().reset_index().rename(columns = {"index" : "Churn","Churn":"Churn_counts"})
inter_calls_churn['percentage'] = inter_calls_churn['Churn_counts'].apply(lambda x : round((x/226)*100,2))
inter_calls_churn
```

```python
#Churn Distribution on basis of call duration type
data_by_duration = new_df.groupby(['Call_duration','Churn'])["Churn"].count().unstack()
data_by_duration = data_by_duration.reset_index()
data_by_duration['total'] = data_by_duration[True] + data_by_duration[False]
data_by_duration['Churn_percentage'] = round((data_by_duration[True]/data_by_duration['total'])*100,2)
data_by_duration.rename(columns = {True:"Churned_true",False:'Churned_false'},inplace = True)
data_by_duration
```

```python
# Churn percentage on the basis of number of customer service calls.
Churn_vs_calls=new_df.groupby('Customer service calls')["Churn"].value_counts()
Churn_vs_calls=Churn_vs_calls.unstack().reset_index().fillna(0)
Churn_vs_calls['percentage_churn']=(Churn_vs_calls[1]/(Churn_vs_calls[1]+Churn_vs_calls[0]))*100
Churn_vs_calls
```

```python
#Distribution of call length on the basis customer service calls
call_duration_vs_customer_support=new_df.groupby(['Call_duration','Customer service calls'])["Customer service calls"].count().unstack().transpose()
call_duration_vs_customer_support=call_duration_vs_customer_support.reset_index()
call_duration_vs_customer_support['long_callers_percentage']=round((((call_duration_vs_customer_support['long'])/280)*100),2)
call_duration_vs_customer_support.head()
```

```python
#checking out relation between voice mails & chrun
voice_mail_churn = new_df[(new_df['Voice mail plan'] == 'Yes')][['Number vmail messages','Churn']]
voice_mail_churn.sort_values(by ='Number vmail messages', ignore_index = True,inplace = True)
voice_mail_churn.head()
```

```python
# Churn percentage among voice mail subscribers
voice_mail_churn = voice_mail_churn['Churn'].value_counts().reset_index().rename(columns = {"index" : "Churn","Churn":"Churn_counts"})
voice_mail_churn['percentage'] = voice_mail_churn['Churn_counts'].apply(lambda x : round((x/922)*100,2))
voice_mail_churn
```

```python
#Finding out total charges in different periods of day
total_charge = new_df[["Total night charge","Total day charge","Total eve charge",'Total intl charge']].sum().reset_index().rename(columns = {'index':'charges',0:'Total_charges'})
total_charge
```

```python
#Finding out total minutes in different periods of day
total_min = new_df[["Total night minutes","Total day minutes","Total eve minutes",'Total intl minutes']].sum().reset_index().rename(columns = {0:"Total_min","index" :'mins'},)
total_min
```

```python
#finding out charges per minute in different duration of day
charges_per_min = pd.concat([total_charge,total_min],axis = 1)
charges_per_min['charges per min'] = round(charges_per_min['Total_charges']/charges_per_min['Total_min'],3)
charges_per_min.drop(['mins','charges'],axis = 1,inplace = True)
charges_df = charges_per_min.rename(index = {0:"Night",1:"Day",2:"Evening",3:"International"}).sort_values(by = 'charges per min',ascending = False)
charges_df = charges_df.reset_index().rename(columns = {"index":"Charge type"})
charges_df
```

```python
#Finding out total charges by state
columns_charge = ["Total night charge","Total day charge","Total eve charge",'Total intl charge']
total_charges = new_df.groupby(['State'])[columns_charge].sum()
total_charges_statewise = total_charges.sum(axis = 1).reset_index().rename(columns={0:"total_charges"}).sort_values(by='total_charges',ascending = False,ignore_index = True)
total_charges_statewise.head()
```

```python
# Finding out total mins by each state.
columns_mins = ['Total night minutes','Total day minutes','Total eve minutes']
total_mins = new_df.groupby(['State'])[columns_mins].sum()
total_mins_statewise = total_mins.sum(axis = 1).reset_index().rename(columns={0:"total_mins"}).sort_values(by='total_mins',ascending = False,ignore_index = True)
total_mins_statewise.head()
```

```python
# Finding out number of churn vs Customer service calls.
chruns_by_states = new_df.groupby(by = ['State'])[['Churn','Customer service calls']].sum().sort_values(by = 'Customer service calls',ascending = False).reset_index()
chruns_by_states.head()
```

```python
# Mergeing total_mins_statewise and chruns_by_states.
merged_min_df = pd.merge(total_mins_statewise,chruns_by_states, on ='State')
merged_df = pd.merge(total_charges_statewise,merged_min_df, on = 'State')
merged_df['charges_per_mins'] = merged_df['total_charges']/merged_df['total_mins']
final_df = pd.merge(merged_df,users_statewise_churn,on = 'State')
final_df.drop(final_df.columns[[3,6,7,8]], axis=1, inplace=True)
final_df.sort_values(by = 'churn percentage',ascending = False,inplace = True)
final_df.head()
```

```python
# Adding new column total_calls. 
columns = ['State','Area code','Total day calls','Total eve calls','Total night calls','Total intl calls']
new_df['total_calls'] = new_df['Total day calls']+new_df['Total eve calls']+new_df['Total night calls']+new_df['Total intl calls']
new_df.groupby('State')['total_calls'].sum().reset_index().head()
```

```python
# dividing Customers and total_calls on the basis area code
customer_by_area_code =  new_df.groupby(['Area code']).agg({'Customer_id': 'count', 'total_calls': 'sum'})
customer_by_area_code
```

```python
# Churn by Area code.
customer_by_churn = new_df.groupby(['Area code','Churn'])['Churn'].count().unstack()
customer_by_churn = customer_by_churn.drop(customer_by_churn.columns[0],axis = 1)
customer_by_churn
```

```python
# churn percentage by area code.
churns_areawise = pd.merge(customer_by_area_code,customer_by_churn,on = 'Area code').reset_index()
churns_areawise.rename(columns = {True:'churns_count','Customer_id':'Number_of_customers'},inplace = True)
churns_areawise['percentage'] = round(100*(churns_areawise['churns_count']/churns_areawise['Number_of_customers']),2)
churns_areawise
```
